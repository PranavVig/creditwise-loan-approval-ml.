# EDA - CreditWise Loan Dataset

This notebook performs basic exploratory analysis.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/raw/loan_applications_sample.csv')
df.head()
```

```python
df.info()
```

```python
df['Loan_Approved'].value_counts(normalize=True)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Credit_Score', hue='Loan_Approved', bins=20, kde=True)
plt.title('Credit Score distribution by Loan Approval')
plt.show()
```
